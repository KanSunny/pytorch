```python
# 手动实现单层全连接层，并训练y = Wfrom IPython import displayx+b
import torch as t
import numpy as np
from torch import nn
from IPython import display
import torch.optim as optim
from matplotlib import pyplot as plt
from torch.autograd import Variable as V

# 获取学习数据
t.manual_seed(1000)

def get_fake_data(batch_size = 8):# 产生列向量８＊１
    x = t.rand(batch_size, 1) * 20
    y = 2*x + (1 + t.randn(batch_size, 1))*3
    return x, y


class Linear(nn.Module):
    def __init__(self):
        super(Linear, self).__init__()
        self.fc1 = nn.Linear(8, 8)
        
    def forward(self, x):
        x = self.fc1(x)
        return x
    
linear = Linear()

optimizer = optim.SGD(linear.parameters(), lr = 0.001)
for i in range(30000):
    # 产生训练数据
    x, y= get_fake_data()
    x, y = V(x.t()), V(y.t())# 转化为行向量１＊８
    # 梯度清零，必须要清零
    optimizer.zero_grad()
    # 计算损失
    y_pred =linear(x)
    criterion = nn.MSELoss()
    loss = criterion(y_pred, y)
    # 反向传播
    loss.backward()
    # 更新参数
    optimizer.step()
    
    if i % 1000 == 0:
        
        # 画图
        display.clear_output(wait = True)
        x = V(t.arange(0, 8).view(1, -1))# 转化为行向量１＊８
        y = linear(x)

        print(x)
        print(y)
        # 不构成一条线的话不能绘制
        plt.plot(x.data.numpy(), y.data.numpy())# predicted 绘制预测的线
        
        x2, y2 = get_fake_data()
        plt.scatter(x2.numpy(), y2.numpy())# true data 绘制数据点
        
        plt.xlim(0, 20)# x轴坐标范围
        plt.ylim(0, 41)# y轴坐标范围
        plt.show()# 程序输出图
        plt.pause(0.5)
```

```python
# 全连接层相当于y = wx+b, y--out_feature, x--in_feature
import torch as t
from torch import nn
from torch.autograd import Variable as V

class Linear(nn.Module):
    def __init__(self, in_feature, out_feature):
        super(Linear, self).__init__()
        # 定义可学习参数
        self.w = nn.Parameter(t.randn(in_feature, out_feature))
        self.b = nn.Parameter(t.randn(out_feature))
        
    def forward(self, x):
        x = x.mm(self.w)
        return x + self.b.expand_as(x)

layer = Linear(4, 4)


optimizer = optim.SGD(layer.parameters(), lr = 0.001)
for i in range(30000):
    # 获取数据
    x, y = get_fake_data(4)
    x, y = V(x.t()), V(y.t())
    # 梯度清零
    optimizer.zero_grad()
    # 计算损失函数
    y_pred = layer(x)
    criterion = nn.MSELoss()
    loss = criterion(y_pred, y)
    # 反向传播
    loss.backward()
    # 更新参数
    optimizer.step()

for name, parameter in layer.named_parameters():
    print(name, parameter)# w, b
x = V(t.arange(0, 4).view(1, -1))# 转化为行向量１＊８
print(x)
y = layer(x)
print(y)
```
