```python
# 加载数据方式1: 使用python魔法方法读取数据, 通过transform修改数据

# 直接加载数据集，与前面使用torchvision下的CIFAR10不一样
# 读取图片数据
import os
import torch as t
import numpy as np
from PIL import Image
from torch.utils import data

class DogCat(data.Dataset):
    def __init__(self, root):
        imgs = os.listdir(root)
        # 所有图片的绝对路径，调用__getitem__才会真正的加载图片
        # root + image name
        self.imgs = [os.path.join(root, img) for img in imgs]
    def __getitem__(self, index):# 返回一张数据和一个样本
        img_path = self.imgs[index]
        # dog->1, cat->0
        # -1 is the last one
        label = 1 if 'dog' in img_path.split('/')[-1] else 0
        pil_img = Image.open(img_path)
        array = np.asarray(pil_img)
        data = t.from_numpy(array)
        return data, label
    def __len__(self):# 返回数据长度
        return len(self.imgs)

dataset = DogCat("/home/kansunny/Public/data/DogsVsCats/")
img, label = dataset[0]# 相当于调用dataset.__getitem__(0)
for img, label in dataset:
    # 形状不一样，图片大小不一样，数据值较大未归一化
    print(img.size(), img.float().mean(), label)
```

```python
# 使用torchvision.transforms进行数据集的初始化
import os
import torch as t
import numpy as np
from PIL import Image
from torch.utils import data
import torchvision.transforms as T

transform = T.Compose([# 为什么一定要compose呢，直接操作每条语句会报错。。。
    T.Resize(224),# 缩放图片
    T.CenterCrop(224),# 从图片中切出224*224的图片
    T.ToTensor(),# 将图片转化为Tensor，归一化为[0, 1]
    T.Normalize([.5, .5, .5], [.5, .5, .5])# 标准化为[-1, -1]
])

class DogCat(data.Dataset):
    
    def __init__(self, root, transforms = None):
        imgs = os.listdir(root)
        # 所有图片的绝对路径，调用__getitem__才会真正的加载图片
        # root + image name
        self.imgs = [os.path.join(root, img) for img in imgs]
        self.transforms = transforms
    
    def __getitem__(self, index):# 返回一张数据和一个样本
        img_path = self.imgs[index]
        # dog->1, cat->0
        # -1 is the last one
        label = 0 if 'dog' in img_path.split('/')[-1] else 1
        data = Image.open(img_path)
        if self.transforms:
            data = self.transforms(data)
        return data, label
    
    def __len__(self):# 返回数据长度
        return len(self.imgs)
    
dataset = DogCat('/home/kansunny/Public/data/DogsVsCats/dog', transforms = transform)

for img, label in dataset:
    print(img.size(), label)
    

data = Image.open("/home/kansunny/Public/data/DogsVsCats/cat/cat.12484.jpg")
transform(data)
```

```python
# 加载数据方式2：使用 ImageFolder 加载数据，使用transform进行数据转化

import torchvision.transforms as T
from torchvision.datasets import ImageFolder

# imagefolder要求每个文件夹内存放相同类别
dataset = ImageFolder('/home/kansunny/Public/data/DogsVsCats/')

# 显示每个文件家对应类别的编号
dataset.class_to_idx 

# 搜索图片的路径和对应的label
dataset.imgs

dataset[0][1]# 第一维表示是第几个图片综合多个文件夹，第二维为1表示显示其label，为0表示显示其数据
dataset[0][0].show()

# 添加transform
transform = T.Compose([
    T.RandomResizedCrop(224),# 切图, 随机切并resize成224
    T.RandomHorizontalFlip(),# 随机水平翻转
    T.ToTensor(),
    T.Normalize(mean = [0.4, 0.4, 0.4], std = [0.2, 0.2, 0.2]),# 正则化，均值方差
])

dataset = ImageFolder('/home/kansunny/Public/data/DogsVsCats/', transform=transform)
dataset[0][0].size()

to_img = T.ToPILImage()
# 0.2 和 0.4 是标准差和均值的近似
to_img(dataset[0][0]*0.2+0.4)
```

```python
# 加载数据之后：使用后使用 DataLoader 可以进行进一步对数据的操作

from torch.utils.data import DataLoader

# 加载数据集 dataset（已经 transform 了），batch_size 为3...
dataloader = DataLoader(dataset, batch_size = 3, shuffle=True, num_workers=0, drop_last=False)
dataiter = iter(dataloader)
imgs, labels = next(dataiter)
imgs.size() 
```
